# RMSF plots

Corresponding to Figures 3 and 5

## Import libraries

```python
import glob
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats

from matplotlib.pyplot import cm
```

## Plot Diff RMSF

```python
def fetch_rmsf(system, first_resid=0, last_resid=-1):
   
    rmsf = []
    for i in sorted(glob.glob(system)):
        b = np.loadtxt(i)[:,1][first_resid:last_resid]
        rmsf.append(b)
        
    ave_rmsf= np.average(rmsf, axis=0)    
    
    return ave_rmsf
        
```

## SERT

```python
SERT_gly = ['nogly', 'gly-N208', 'gly-N217',  'gly-N208-N217']
path = "calculated_features/SERT/rmsf/" 
color_map = cm.rainbow(np.linspace(0, 1, len(SERT_gly)))

first_resid, last_resid = [0, 542]

rmsf = []

for condition in SERT_gly:    
    system = f"{path}/{condition}/SERT-{condition}-5HT-NaCl-*-rmsf.dat"
      
    ave_rmsf = fetch_rmsf(system, first_resid, last_resid)
    
    rmsf.append(ave_rmsf)

    
plt.figure(figsize=(13,4))
for i,color in zip(range(len(rmsf)), color_map):
    plt.plot(rmsf[i]-rmsf[0], c=color)
    

plt.plot([132,132], [-2, 2], linestyle="--", c='k', linewidth=0.5, zorder=-1)
plt.plot([141,141], [-2, 2], linestyle="--", c='k', linewidth=0.5, zorder=-1)

plt.axvspan(9,36, color='darkgray', alpha=0.2)

plt.axvspan(39,67, color='darkgray', alpha=0.2)
plt.axvspan(80,115, color='darkgray', alpha=0.2)

plt.axvspan(177,195, color='darkgray', alpha=0.2)
plt.axvspan(198,225, color='darkgray', alpha=0.2)

plt.axvspan(247,274, color='darkgray', alpha=0.2) 

plt.axvspan(281,313, color='darkgray', alpha=0.2)
plt.axvspan(344,377, color='darkgray', alpha=0.2)

plt.axvspan(386,405, color='darkgray', alpha=0.2)
plt.axvspan(408,440, color='darkgray', alpha=0.2)

plt.axvspan(459,482, color='darkgray', alpha=0.2)
plt.axvspan(496,523, color='darkgray', alpha=0.2)

plt.xticks(np.arange(first_resid,last_resid+1,50), np.arange(first_resid,last_resid+1,50)+76, fontsize=12)

plt.xlabel("Residue No.")
plt.xlim([first_resid, last_resid])

plt.ylabel("RMSF")
plt.ylim([-2,2])



#plt.savefig('SERT-RMSF-diff.svg', dpi=300)
```
