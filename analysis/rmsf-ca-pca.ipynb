## Import libraries

```python
import numpy as np
import glob
from matplotlib import pyplot as plt
from scipy import stats
from scipy.signal import savgol_filter
import seaborn as sns
from matplotlib.pyplot import cm
from sklearn.decomposition import PCA
```

## fetch rmsf data

```python
def fetch_rmsf(system_path, condition, color, first_resid=0, last_resid=-1):
    rmsf = []
    color_out = []
    for i in sorted(glob.glob(system_path)):
        b = np.loadtxt(i)[:,1][first_resid:last_resid]
        rmsf.append(b)
        color_out.append(color)
    return rmsf, color_out
```

## SERT PCA

```python
gly_cond = ['nogly', 'gly-N208', 'gly-N217',  'gly-N208-N217']
path = "calculated_features/SERT/rmsf-ca/" 
color_map = cm.rainbow(np.linspace(0, 1, len(gly_cond)))
first_resid, last_resid = [0, 542]

all_rmsf = []
color_plot = []
for condition,color in zip(gly_cond, color_map):    
    system = f"{path}/{condition}/SERT-{condition}-5HT-NaCl-CLONE*rmsf.dat"
    rmsf, color_out = fetch_rmsf(system, condition, color, first_resid, last_resid)
    
    all_rmsf.extend(rmsf)
    color_plot.extend(color_out)

plt.figure(figsize=(4,4))
pca = PCA(n_components=2)
comps = pca.fit_transform(all_rmsf)
plt.scatter(comps[:,0], comps[:,1], c=color_plot)
plt.show()
#plt.savefig("SERT-pca.svg")
```
